```python
import matplotlib.pyplot as plt

def quadratic_temperature_model(day, a, b, c):
    # Quadratic equation for temperature
    temperature = a * day**2 + b * day + c
    return temperature

def main():
    # Fixed coefficients for the quadratic equation: ax^2 + bx + c
    a = 5
    b = 3
    c = 4

    # Fixed number of days to model
    num_days = int(input("Enter the number of days to model: "))

    # Lists to store day and corresponding temperature values
    days = list(range(1, num_days + 1))
    temperatures = [quadratic_temperature_model(day, a, b, c) for day in days]

    # Plotting
    plt.plot(days, temperatures, label=f'Temperature: {a}x^2 + {b}x + {c}')
    plt.title('Weather Modeling')
    plt.xlabel('Day')
    plt.ylabel('Temperature (°C)')
    plt.legend()
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_temperature_model(time, a, b, c):
    """
    Quadratic temperature model: T(time) = a*time^2 + b*time + c
    """
    return a * time**2 + b * time + c

def plot_temperature_model(a, b, c):
    """
    Plot the quadratic temperature model for a specific day.
    """
    time_values = np.linspace(0, 24, 100)  # 24 hours in a day
    temperature_values = quadratic_temperature_model(time_values, a, b, c)

    plt.figure(figsize=(10, 6))
    plt.plot(time_values, temperature_values, label='Temperature Model', color='blue')
    plt.xlabel('Time (hours)')
    plt.ylabel('Temperature')
    plt.title('Quadratic Temperature Model for a Day')
    plt.legend()
    plt.grid(True)
    plt.show()

def main():
    # Hard-coded coefficients for the quadratic equation
    a_fixed = 0.02
    b_fixed = 1.5
    c_fixed = 20

    # Get user input for coefficients (you can customize this part)
    a_user = float(input("Enter a coefficient for time^2: "))
    b_user = float(input("Enter a coefficient for time: "))
    c_user = float(input("Enter a constant term: "))

    # Combine fixed and user coefficients
    a_combined = a_fixed + a_user
    b_combined = b_fixed + b_user
    c_combined = c_fixed + c_user

    # Plot the temperature model
    plot_temperature_model(a_combined, b_combined, c_combined)

if __name__ == "__main__":
    main()
```
